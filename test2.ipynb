```python
st='Print only the words that start with s in this sentence'
```

```python
splitting=st.split(' ')      #splits in form of lists. Splir by default splits on white space
print(splitting)
for i in splitting:
    if i[0]=='s' or i[0]='S':     #i[0].lower()=='s'. Forces the capital S letter to be lower case and in this it will grab it and we dont have to apply or statement
        print (i)
```

```python
st='Print every word in this sentence that has an even number of letters'
```

```python
splitted=st.split(' ')
print(splitted)
for i in splitted:
    if (len(i)%2==0): 
        print(i)
```

```python
num1=range(1,101)
```

```python
for i in num1:
    if (i%3==0 and i%5==0):
        print ('FizzBuzz')
    elif (i%3==0):
        print ('Fizz')
    elif (i%5==0):
        print ('Buzz')
    else:
        print (i)
```
